```python
DATA_PATH = '../../data/'

LEAGUE = "UFC"
SEASON = "2020"

# Tags: Fighter stats, Fighters, Standings
SD_API_FIGHTERS_EP = "https://api.sportsdata.io/v3/mma/scores/json/Fighters"

#Tags: Fight Day Info, Results, Schedules
SD_API_SCHEDULE_BASE = "https://api.sportsdata.io/v3/mma/scores/json/Schedule/"
SD_API_SCHEDULE_EP = SD_API_SCHEDULE_BASE + "/" + LEAGUE + "/" + SEASON

# Tags: Fight Day Info, Results
SD_API_EVENT_EP = "https://api.sportsdata.io/v3/mma/scores/json/Event/"

# Tags: Fight Day Info, Fight State, Results
SD_API_FIGHT_EP = "https://api.sportsdata.io/v3/mma/stats/json/Fight/"

# Tags: Fighter stats, Fighters, Standings
SD_API_FIGHTER_EP = "https://api.sportsdata.io/v3/mma/scores/json/Fighter/"
```

```python
format_dict = {'format':'json'}
sched_dict = {'league':LEAGUE, 'season':SEASON}
```

```python
session = requests.Session()
session.headers.update({'Ocp-Apim-Subscription-Key' : 'c0cef4085d8444c98c5a2614c96e436e'})
```

```python
# merge dictionaries for schedule api query
sched_query = {**format_dict, **sched_dict} 

sched_response = session.get(SD_API_SCHEDULE_EP, params=sched_query)
```

```python
events_json = []

for event in sched_response.json():
    eventid = event['EventId']
    event_dict = {'eventid':eventid}
    
    # merge dictionaries for event api query
    event_query = {**format_dict, **event_dict} 
    event_response = session.get(SD_API_EVENT_EP + "/" + str(eventid), params=event_query)
    events_json.append(event_response.json())
```

```python
events_json
```

```python
fight_json = []

# the JSON object returned from the events request has all events for the year
for events in events_json:
    # iterate through each fight for the current event
    for fight in events['Fights']:
        fightid = fight['FightId']
        fight_dict = {'fightid':fightid}

        # merge dictionaries for event api query
        fight_query = {**format_dict, **fight_dict} 
        fight_response = session.get(SD_API_FIGHT_EP + "/" + str(fightid), params=fight_query)
        fight_json.append(fight_response.json())
```

```python
fight_json
```

```python
fighters_list = []

# the JSON object returned from the events request has all events for the year
for events in events_json:
    # iterate through each fight for the current event
    for fight in events['Fights']:
        for fighters in fight['Fighters']:
            fighters_list.append(fighters['FighterId'])
            
# we only need each fighter represented once
fighters_set = set(fighters_list)
```

```python
fighter_json = []

for fighterid in fighters_set:
    fighter_dict = {'fighterid':fighterid}

    # merge dictionaries for event api query
    fighter_query = {**format_dict, **fighter_dict} 
    fighter_response = session.get(SD_API_FIGHTER_EP + "/" + str(fighterid), params=fighter_query)
    fighter_json.append(fighter_response.json())
```

```python
fighter_json
```

```python
fighter_df = pd.DataFrame(data=fighter_json)
fighter_df.head()
```

    We see that career stats contains a dict/json object with some additional fighter info, we want this directly in our dataframe

```python
career_stats_df = pd.DataFrame(data=fighter_df['CareerStats'].iloc[0], index=[0])
career_stats_df

for CareerStats in fighter_df['CareerStats'].iloc[1:]:
    career_stats_df = career_stats_df.append(CareerStats, ignore_index=True)
    
career_stats_df.head()
```

```python
print(fighter_df.info())
print(career_stats_df.info())
```

```python
# drop the duplicate columns, and column_stats
fighter_df = fighter_df.drop(columns=['CareerStats'])

career_stats_df = career_stats_df.drop(columns=['FighterId', 'FirstName', 'LastName'])
```

```python
# since the career_stats dataframe was taken from the figher dataframe we can use index to join
fighter_stats_df = fighter_df.join(career_stats_df)

# transpose so we can see all columns
fighter_stats_df.head().transpose()
```

```python
fighter_stats_df.to_csv(DATA_PATH + "fighter_stats_" + LEAGUE + "_" + SEASON + ".csv", index=False)
```
